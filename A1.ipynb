```python
#Top 10 countries with highest total percentage of internet access level
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

colors = ['#00a7e4','#00b8e5','#00c6d7','#00d2b9','#00db92','#77e065','#d0de2d','#ecdb1d','#ffd816','#ffe050']
bar221 = sns.barplot(data=top_10_countries, x="Countries and areas", y="Total(%)", palette = colors, ci = None,linewidth=10)
plt.title("Figure2.2.1:", size=14)
bar221.tick_params(axis='x', size = 5,rotation=90)
for p in bar221.patches:
    bar221.annotate(round(p.get_height(),2),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='top',
                   xytext=(0, 9), size= 12, color = 'blue',
                   textcoords='offset points')
```

#### The country with the highest percentage of internet access for students between the ages of 3-17 is Serbia, with a rate of 94%. North Macedonia and Tonga follow with the second and third-highest rates, respectively.

```python
import pandas as pd
import matplotlib.ticker as mtick

colors = ['#00a7e4','#ffe050']
bar222 = top_10_countries.pivot(index=["Countries and areas"], columns='Income Group', values='Total(%)').plot(kind='bar',color =colors,title ="Figure2.2.2:", stacked=True)
bar222.set_ylabel("Total(%)")
bar222.legend(bbox_to_anchor=(1, 1.02), loc='upper left', title = "Income Group")
for container in bar222.containers:
    bar222.bar_label(container)
    bar222.yaxis.set_major_formatter(mtick.PercentFormatter())
```

#### We can see from the chart that most of the countries in the top 10 countries with highest internet access level in 3-17 years old students are from upper middle-income countries. Only few are from high-income countries. This suggests that there is a correlation between income level and internet access level, with higher income levels indicating better internet access.
